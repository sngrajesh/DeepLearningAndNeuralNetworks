```python
###----------------
### Import Libraries
###----------------
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt


from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

##### Hyperparameters


```python
###----------------
#Global Variables
###----------------

TEST_SIZE = 0.2
RANDOM_STATE = 24
np.random.seed(RANDOM_STATE)

# set location of input files
inpDir = '../input'

# set location of output files
outDir = '../output'

# Set plotting parameters
params = {
    'figure.figsize': (15,4),
    'axes.titlesize': 'large',
    'axes.labelsize': 'medium',
    'legend.fontsize': 'medium',
    'xtick.labelsize': 'medium',
    'ytick.labelsize': 'medium',
}
plt.rcParams.update(params)


CMAP = plt.cm.coolwarm
plt.style.use('seaborn-v0_8-darkgrid')
```

##### Splitting the data into training and test sets

```python
X = data_df.drop(['target'], axis=1)
y = data_df['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=TEST_SIZE, random_state=RANDOM_STATE, stratify=y)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
X_train['total'] = X_train.astype(np.int64).sum(axis=1)
X_train['target'] = y_train
X_train.sort_values('total', inplace=True)
X_train.head()
```

```python
X_train.tail()
```

```python
X_test['total'] = X_test.astype(np.int64).sum(axis=1)
X_test['target'] = y_test
X_test.sort_values('total', inplace=True)
X_test.head()
```

```python
accuracy = {}

for b in range(X_train.shape[1]):
    X_train['pred'] = X_train['total'] >= b
    accuracy[b] = accuracy_score(y_true=X_train['target'], y_pred=X_train['pred'])
                                 
accuracy
```

##### Getting max threshold value

```python
best_b = max(accuracy, key=accuracy.get)
best_b, accuracy[best_b]
```

##### Predinting on testing data

```python
X_test['pred'] = X_test['total'] >= best_b

accuracy_score(y_true=X_test['target'], y_pred=X_test['pred'])
```
